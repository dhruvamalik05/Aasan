```python
input_shape = len(train_normalized.iloc[0, :])
input_shape
```

```python
dense_nn = tf.keras.Sequential([
            tf.keras.Input(shape=input_shape),
            tf.keras.layers.Dense(100, activation="relu"),
            tf.keras.layers.Dense(100, activation="relu"),
            tf.keras.layers.Dense(100, activation="relu"),
            tf.keras.layers.Dense(2, activation="softmax")
    ])
```

```python
dense_nn.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
train_normalized = tf.convert_to_tensor(train_normalized)
train_normalized
```

```python
train_labels[train_labels == "bad"] = 0.0
```

```python
train_labels[train_labels == "good"] = 1.0
```

```python
train_labels
```

```python
test_labels[test_labels == "bad"] = 0.0
test_labels[test_labels == "good"] = 1.0
```

```python
train_labels_1 = train_labels.to_numpy().astype("float64")
train_labels_1
```

```python
test_labels_1 = test_labels.to_numpy().astype("float64")
test_labels_1
```

```python
train_normalized
```

```python
dense = dense_nn.fit(train_normalized, train_labels_1, epochs=10, validation_data=(test_normalized, test_labels_1))
```
