```python
import numpy as np
```

```python
from sklearn.metrics import classification_report
```

```python
import tensorflow as tf
import os
```

```python
from skimage import io
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
count = 1
for file in os.listdir("test videos\\bad"):
    print(file)
    cap = cv2.VideoCapture("test videos\\bad\\"+file)
    while True:
        success, img = cap.read()
        print(success)
        if success == False:
            break
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        cv2.imwrite("test videos\\bad_frames\\" + str(count) + ".png", cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        print(count)
        count +=1
```

```python
path = "train videos\\bad_frames"
training = []
for file in os.listdir(path):
    #print(path+"\\"+file)
    img = cv2.imread(path+"\\"+file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    new_img = cv2.resize(img, (352, 640))
    training.append([new_img, 0])
#     plt.imshow(new_img)
#     plt.show()
#     print(new_img)
```

```python
path = "train videos\\good_frames"
for file in os.listdir(path):
    #print(path+"\\"+file)
    img = cv2.imread(path+"\\"+file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    new_img = cv2.resize(img, (352, 640))
    training.append([new_img, 1])
#     plt.imshow(new_img)
#     plt.show()
#     print(new_img)
```

```python
print(len(training[0]))
```

```python
import random
random.shuffle(training)
```

```python
training = np.array(training)
print(training.shape)
```

```python
training
```

```python
X = []
Y = []
for features, label in training:
    X.append(features)
    Y.append(label)
```

```python
X = np.array(X)
X = X.astype("float32")
```

```python
X /= 255
```

```python
input_shape = X[0].shape
input_shape
```

```python
Y = [float(i) for i in Y]
```

```python
Y = np.array(Y)
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=input_shape),
#     tf.keras.layers.Conv2D(16, (3,3), activation="relu"),
#     tf.keras.layers.MaxPooling2D((3, 3)),
    tf.keras.layers.Conv2D(32, (3,3), activation="relu"),
    tf.keras.layers.MaxPooling2D((3, 3)),
    tf.keras.layers.Conv2D(32, (3,3), activation="relu"),
    tf.keras.layers.MaxPooling2D((3, 3)),
    tf.keras.layers.Conv2D(64, (3,3), activation="relu"),
    tf.keras.layers.MaxPooling2D((3, 3)),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dense(2,  activation="softmax")
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(X, Y, batch_size = 16, epochs = 1, verbose = 1)
```

```python
model.save("cnn")
```
